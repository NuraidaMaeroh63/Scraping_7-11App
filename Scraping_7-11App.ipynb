# Get App 

```python
# Import module ต่างๆเพื่อนำไปใช้ต่อไปได้
import pandas as pd
import numpy as np
import json
from tqdm.auto import tqdm

#เข้าถึง AppStore เพื่อดึงข้อมูล  
from app_store_scraper import AppStore
seven = AppStore(country='th', app_name='7-Eleven TH', app_id = '514262377')

seven.review(how_many=2000)
```

```python
#ดูข้อมูล
seven.reviews
```

# Step 3 – Convert Data from JSON

```python
# ทำให้อยู่ในรูป Data Frame
sevendf = pd.DataFrame(np.array(seven.reviews),columns=['review'])
sevendf2 = sevendf.join(pd.DataFrame(sevendf.pop('review').tolist()))

sevendf2.head()
```

```python
#Drop Column ที่ไม่ได้นำไปใช้ต่อ
sevendf2 = sevendf2.drop('isEdited' , axis = 1)
sevendf2 = sevendf2.drop('developerResponse' , axis = 1)
```

```python
sevendf2.head()
```

```python
#ตรวจ type ให้อยู่ในมาตรฐาน
sevendf2.dtypes
```

# Step 4 – Convert the Dataframe to CSV

```python
#จาก Data Frame แปลงให้อยู่ในรูป C
sevendf2.to_csv('7-Eleven TH-app-reviews.csv', index = False)
```
